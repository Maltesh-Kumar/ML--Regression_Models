## Simple Regression Equation: y = m*x+ c

```python
m = 0
c = 0
y = lambda x : m*x+c
```

```python
def plot_line(y, data_points):
    x_val = [i for i in range(int(min(data_points))-1, int(max(data_points))+2)]
    y_val = [y(x) for x in x_val]
    plt.plot(x_val, y_val, 'r')
```

```python
plot_line(y, x_points)
plt.grid(alpha = 0.3)
```

## This is initial best fit line

```python
learn = 0.001
```

## Gradient Descent Formula

```python
def summation(y, x_points, y_points):
    total_1 = 0
    total_2 = 0
    
    for i in range(1,len(x_points)):
        total_1 += y(x_points[i]) - y_points[i]
        total_2 += (y(x_points[i]) - y_points[i])*(x_points[i])
        
    return total_1/len(x_points),total_2/len(x_points)
```

```python
for i in range(1000):
    s1,s2 = summation(y, x_points, y_points)
    m = m-learn*s2
    c = c-learn*s1
    plot_line(y, x_points)
    plt.plot(x_points, y_points, 'bo')
    plt.grid(alpha = 0.7) 
```

## This shows how the line started from horizontal line with maximum errors and then later moved towards least possible error
